```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math
#300km square area
#3000 time steps or 3000seconds or 50 minutes? 400 Objects? 6 waypoints each sim
#min distance bw waypoints is 10km
#need to do random selection of tower


#create group of origins and group of ending waypoints? max field is 300km by 400km?
origins = [[0,20],[1000,1000],[3000,10000],[10000,3500],[4800,5000],[2000,3000],[4100,800],[7800,2300]]
endings = [[180000,220000],[295000,398000],[245000,310000]]

#set min dist for waypoint
min_dist = 20000

#for loop of 400 planes:

#choose origin from list
origin = random.choice(origins)
first_path = []
first_path.append(origin)
#choose ending from list
ending = random.choice(endings)
#specify 5 or 6 random waypoints (each new waypoint must satisfy requirements)
wp_ptr = origin
for i in range (0,5):
    x = random.randrange(origin[0],ending[0])
    y = random.randrange(origin[1],ending[1])
    #requirements: new wp closer to ending than prev wp, min & max distance bw wp's
    while math.sqrt((ending[0]-x)**2 + (ending[1]-y)**2) > math.sqrt((ending[0]-wp_ptr[0])**2 + (ending[1]-wp_ptr[1])**2):
        x = random.randrange(origin[0],ending[0])
        y = random.randrange(origin[1],ending[1]) 
    wp_ptr = [x,y]
    first_path.append([x,y])
    #save in matrix? struct?
first_path.append(ending)
print(origin)
print(ending)
print(first_path)
```

```python
x_list = []
y_list = []
for coord in first_path:
    x_list.append(coord[0])
    y_list.append(coord[1])


plt.plot(x_list,y_list)
plt.show()
```
